```python
import networkx as nx
X = nx.read_graphml(file_path)
```

Once we have the DiGraph object X in memory, let's compute the in- and out-degree sequences of the Digraph...

```python
in_degrees = X.in_degree()  # Returns a DegreeView object with node and in-degree
out_degrees = X.out_degree()  # Returns a DegreeView object with node and out-degree

# To get the degree sequences as lists:
in_degree_sequence = [d for n, d in in_degrees]
out_degree_sequence = [d for n, d in out_degrees]
```

Let's plot two histograms for the values of the in- and out-degrees...

Notice how most people have low in- and out-degrees, but the histogram is very skew towards high degrees (similar to the distribution of wealth)...

When the range of values of a variable (degree) is very large, it is convenient to plot the histogram in log-log scale...

```python
import numpy as np

# Function to compute the histogram with logarithmic bins
def log_binning(sequence):
    # Filter out zero-degree nodes
    filtered_sequence = [deg for deg in sequence if deg > 0]

    # Check if the filtered sequence is empty or has a single element
    if not filtered_sequence or len(filtered_sequence) == 1:
        return np.array([]), np.array([])

    # Compute histogram with logarithmic bins
    hist, edges = np.histogram(filtered_sequence, bins=np.logspace(np.log10(min(filtered_sequence)), np.log10(max(filtered_sequence)), num=20))
    return hist, edges

# Compute histograms for in-degree and out-degree sequences
in_hist, in_edges = log_binning(in_degree_sequence)
out_hist, out_edges = log_binning(out_degree_sequence)

# Plotting
plt.figure(figsize=(12, 6))

# Function to plot a log-log bar histogram
def plot_loglog_bar(hist, edges, subplot_index, title, color):
    if hist.size and edges[:-1].size:
        plt.subplot(1, 2, subplot_index)
        # Calculate bar widths and positions
        widths = np.diff(edges)
        positions = edges[:-1]
        plt.bar(positions, hist, width=widths, color=color, edgecolor='black', align='edge', log=True)
        plt.xscale('log')
        plt.title(title)
        plt.xlabel('Log of Degree')
        plt.ylabel('Log of Frequency')

# Plotting in-degree histogram
plot_loglog_bar(in_hist, in_edges, 1, 'In-Degree Histogram (Log-Log scale)', 'blue')

# Plotting out-degree histogram
plot_loglog_bar(out_hist, out_edges, 2, 'Out-Degree Histogram (Log-Log scale)', 'green')

plt.tight_layout()
plt.show()
```

It is also common to plot a scatter plot of the Degrees vs the Frequencies in log-log scale...

```python
# Function to calculate degree frequencies
def calculate_frequencies(degree_sequence):
    # Count the frequency of each degree
    degree_counts = Counter(degree_sequence)
    # Filter out zero degrees
    filtered_degrees = {deg: count for deg, count in degree_counts.items() if deg > 0}
    degrees = np.array(list(filtered_degrees.keys()))
    frequencies = np.array(list(filtered_degrees.values()))
    return degrees, frequencies

# Calculate frequencies for in-degrees and out-degrees
in_degrees, in_frequencies = calculate_frequencies(in_degree_sequence)
out_degrees, out_frequencies = calculate_frequencies(out_degree_sequence)

# Plotting
plt.figure(figsize=(12, 6))

# Scatter plot for in-degrees
plt.subplot(1, 2, 1)
plt.scatter(in_degrees, in_frequencies, color='blue')
plt.xscale('log')
plt.yscale('log')
plt.title('In-Degree vs Frequency')
plt.xlabel('In-Degree (Log Scale)')
plt.ylabel('Frequency (Log Scale)')

# Scatter plot for out-degrees
plt.subplot(1, 2, 2)
plt.scatter(out_degrees, out_frequencies, color='green')
plt.xscale('log')
plt.yscale('log')
plt.title('Out-Degree vs Frequency')
plt.xlabel('Out-Degree (Log Scale)')
plt.ylabel('Frequency (Log Scale)')

plt.tight_layout()
plt.show()
```
